```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("apjspeech.pdf")
docs=loader.load_and_split()
docs
```

```python
from langchain_core.prompts import PromptTemplate

template = """
Write a concise and short summary of the following speech:

Speech:
{text}
"""

prompt = PromptTemplate(
    input_variables=['text'],
    template=template
)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain=load_summarize_chain(llm,chain_type="stuff",prompt=prompt,verbose=True)
chain.run(docs)
```

```python
## MAP REDUCE
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_spilter=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100)

final_split_docs=text_spilter.split_documents(docs)
```

```python
final_split_docs
```

```python
len(final_split_docs)
```

```python
from langchain_core.prompts import PromptTemplate

template = """
Write a concise and short summary of the following speech:

Speech:
{text}
"""

prompt = PromptTemplate(
    input_variables=['text'],
    template=template
)
```

```python
final_prompt="""  

proivde the final summary in these following ways 
start with motivation quotes and start with precise 
summary and introduction and proivde full understandable summary 
speech:{text}

"""

final_prompt_template=PromptTemplate(input_variables=['text'],template=final_prompt)
```

```python
from langchain.chains.summarize import load_summarize_chain

chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=prompt,      # used on each chunk
    combine_prompt=final_prompt_template,                 # used to combine summaries
    verbose=True
)
chain.run(final_split_docs)
```

```python
## REFINE CHAIN
```

```python
chain=load_summarize_chain(llm,chain_type="refine",verbose=True)
chain.run(final_split_docs)
```
